## And we plot the comparison : 

```python
#OSPA
ospa_ = [ospa1,ospa2,ospa3]
estimated_ = [estimated1,estimated2,estimated3]
labels_ = [r'PHD with $\rho$={} and J={}'.format(rho_[i], J_[i]) for i in range(3)]

cols = ['k','r','b']
plt.figure(figsize=(15,6))

for i,(ospa,col) in enumerate(zip(ospa_,cols)):
    plt.plot(ospa.mean(axis=0), c=col, label=labels_[i], linewidth=4)
    plt.plot(ospa.mean(axis=0) + ospa.std(axis=0), '--'+col, label=labels_[i] + ' +-sd')
    plt.plot(ospa.mean(axis=0) - ospa.std(axis=0), '--'+col, label=labels_[i] + ' +-sd')

    
handles, labels = plt.gca().get_legend_handles_labels()
by_label = OrderedDict(zip(labels, handles))
plt.title('OSPA metric mean over 50 predictions', fontsize=18)
plt.legend(by_label.values(), by_label.keys(), bbox_to_anchor = [1.65,1], fontsize = 25);

#Number of Targets
plt.figure(figsize=(15,6))

plt.plot(true1[0], 'b', linewidth=2, label="True number of targets")

for i,(estimated,col) in enumerate(zip(estimated_,cols)):

    plt.plot(estimated.mean(axis=0), c=col, label= labels_[i], linewidth=4)
    plt.plot(estimated.mean(axis=0) + estimated.std(axis=0), '--'+col, label= labels_[i]+" +-sd")
    plt.plot(estimated.mean(axis=0) - estimated.std(axis=0), '--'+col, label= labels_[i]+" +-sd")

plt.title('Mean estimated number of targets over 50 predictions', fontsize=18)
handles, labels = plt.gca().get_legend_handles_labels()
by_label = OrderedDict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), bbox_to_anchor = [1.5,1], fontsize=15)
plt.show();
```
